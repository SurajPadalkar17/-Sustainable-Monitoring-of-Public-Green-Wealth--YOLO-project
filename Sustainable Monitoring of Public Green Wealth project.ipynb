```python
from google.colab import drive
drive.mount('/content/drive')

import zipfile

# Extract your dataset
zip_path = "/content/drive/MyDrive/Colab Notebooks/project/archive (3).zip"
extract_to = "/content/drive/MyDrive/Colab Notebooks/project/archivee"

with zipfile.ZipFile(zip_path, 'r') as zip_ref:
    zip_ref.extractall(extract_to)

print("✅ Dataset extracted!")
```

```python
yaml_content = """
train: /content/drive/MyDrive/Colab Notebooks/project/archivee/train/images
val: /content/drive/MyDrive/Colab Notebooks/project/archivee/valid/images

nc: 1
names:
  - tree
"""

with open('/content/drive/MyDrive/Colab Notebooks/project/archivee/data.yaml', 'w') as f:
    f.write(yaml_content)

print("✅ data.yaml created!")

```

```python
pip install ultralytics
```

```python
from ultralytics import YOLO

# Load base model (use yolov8s for better accuracy)
model = YOLO("yolov8s.pt")

# Train the model
model.train(
    data='/content/drive/MyDrive/Colab Notebooks/project/archivee/data.yaml',
    epochs=100,
    imgsz=640,
    batch=16,
    patience=20,
    name='tree_growth_detector_v2',
    optimizer='Adam',
    project='/content/drive/MyDrive/Colab Notebooks/project/',
    augment=True
)
```

```python
metrics = model.val()
print("✅ Evaluation complete")
```
